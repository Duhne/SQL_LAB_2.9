# This is how to bring full solutions from MySQL

```python
# -- 1. How many distinct (different) actors' last names are there?
data1 = pd.read_sql_query(''' SELECT distinct last_name, first_name, actor_id FROM sakila.actor''', engine)
data1
```

```python
# -- 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
data2 = pd.read_sql_query('''SELECT *, 
CASE
WHEN weekday(rental_date) then 'weekday'
ELSE 'weekend'
END AS 'day_type'
FROM sakila.rental''', engine)
data2
```

```python
# -- 3. Get all films with ARMAGEDDON in the title
data3 = pd.read_sql_query('''
SELECT * FROM sakila.film
WHERE title LIKE "%%ARMAGEDDON%%" ''', engine)
data3
```

```python
# -- 4. Get 10 the longest films.
data4 = pd.read_sql_query(''' 
SELECT film_id, title, length AS 'Longest 10 Films' FROM sakila.film
ORDER BY length DESC
''', engine)
data4.head(10)
```

```python
# -- 5. How many films include Behind the Scenes content?
data5 = pd.read_sql_query(''' 
SELECT COUNT(special_features) FROM sakila.film
WHERE special_features LIKE "%%Behind the Scenes%%"
''', engine)
data5
```

```python
# -- 6. Which kind of movies (rating) have a mean duration of more than two hours?
data6 = pd.read_sql_query(''' 
SELECT rating, round(AVG(length),2) as 'average_legnth' FROM sakila.film
WHERE length > 120
Group by rating
''', engine)
data6
```

```python
# -- 7. Rank films by length (filter out the rows that have nulls or 0s in length column).
# In your output, only select the columns title, length, and the rank.
data7 = pd.read_sql_query(''' 
SELECT title, rating, length FROM sakila.film
WHERE length != '' OR null
ORDER BY length DESC;''', engine)
data7
```
